```python
import pandas as pd
import numpy as np

print("Libraries loaded successfully")
```

## 1. Load ETH and BTC Data

```python
# Load ETH data (all columns)
eth_df = pd.read_csv('5s_data/ETHUSD_5s.csv')

print(f"ETH data loaded: {eth_df.shape}")
print(f"ETH columns: {list(eth_df.columns)}")
print(f"\nFirst few rows:")
eth_df.head()
```

```python
# Load BTC data (all columns)
btc_df = pd.read_csv('5s_data/BTCUSD_5s.csv')

print(f"BTC data loaded: {btc_df.shape}")
print(f"BTC columns: {list(btc_df.columns)}")
print(f"\nFirst few rows:")
btc_df.head()
```

## 2. Add Prefixes to Column Names

```python
# Add ETH_ prefix to all ETH columns except timestamp
eth_renamed = eth_df.copy()
eth_cols_to_rename = {col: f'ETH_{col}' for col in eth_df.columns if col != 'timestamp'}
eth_renamed = eth_renamed.rename(columns=eth_cols_to_rename)

print(f"ETH columns after renaming: {list(eth_renamed.columns)}")

# Add BTC_ prefix to all BTC columns except timestamp
btc_renamed = btc_df.copy()
btc_cols_to_rename = {col: f'BTC_{col}' for col in btc_df.columns if col != 'timestamp'}
btc_renamed = btc_renamed.rename(columns=btc_cols_to_rename)

print(f"\nBTC columns after renaming: {list(btc_renamed.columns)}")
print(f"\nETH shape: {eth_renamed.shape}")
print(f"BTC shape: {btc_renamed.shape}")
```

## 3. Convert Timestamps

```python
# Convert timestamps to datetime for proper merging
eth_renamed['timestamp'] = pd.to_datetime(eth_renamed['timestamp'])
btc_renamed['timestamp'] = pd.to_datetime(btc_renamed['timestamp'])

print(f"ETH timestamp range: {eth_renamed['timestamp'].min()} to {eth_renamed['timestamp'].max()}")
print(f"BTC timestamp range: {btc_renamed['timestamp'].min()} to {btc_renamed['timestamp'].max()}")
print(f"\nETH rows: {len(eth_renamed):,}")
print(f"BTC rows: {len(btc_renamed):,}")
```

## 4. Merge Data on Timestamp

Merge both datasets using merge_asof to match nearest timestamps within 1 second.

```python
# Sort both dataframes by timestamp (required for merge_asof)
eth_renamed = eth_renamed.sort_values('timestamp').reset_index(drop=True)
btc_renamed = btc_renamed.sort_values('timestamp').reset_index(drop=True)

# Merge using merge_asof to find nearest matching timestamps within 1 second
combined_df = pd.merge_asof(
    eth_renamed,
    btc_renamed,
    on='timestamp',
    direction='nearest',  # Find nearest match (forward or backward)
    tolerance=pd.Timedelta('1s')  # Within 1 second
)

print(f"Combined data shape: {combined_df.shape}")
print(f"Columns: {len(combined_df.columns)} total")
print(f"\nColumn names:")
for i, col in enumerate(combined_df.columns, 1):
    print(f"  {i}. {col}")

print(f"\nFirst few rows:")
combined_df.head()
```

## 5. Check Data Quality

```python
# Check for missing values in BTC columns
btc_cols = [col for col in combined_df.columns if col.startswith('BTC_')]
print(f"Checking {len(btc_cols)} BTC columns for missing data...\n")

for col in btc_cols[:5]:  # Show first 5 BTC columns as examples
    missing = combined_df[col].isna().sum()
    pct = 100 * missing / len(combined_df)
    print(f"  {col}: {missing:,} missing ({pct:.2f}%)")

if len(btc_cols) > 5:
    print(f"  ... and {len(btc_cols) - 5} more BTC columns")


total_btc_missing = combined_df[btc_cols].isna().any(axis=1).sum()
print(f"\nRows with any missing BTC data: {total_btc_missing:,} / {len(combined_df):,} ({100*total_btc_missing/len(combined_df):.2f}%)")
```

## 6. Data Quality Check

```python
# Summary statistics
eth_cols = [col for col in combined_df.columns if col.startswith('ETH_')]
btc_cols = [col for col in combined_df.columns if col.startswith('BTC_')]

print("Combined Dataset Summary:")
print(f"  Total rows: {len(combined_df):,}")
print(f"  Total columns: {len(combined_df.columns)}")
print(f"    - Timestamp: 1")
print(f"    - ETH columns: {len(eth_cols)}")
print(f"    - BTC columns: {len(btc_cols)}")
print(f"  Date range: {combined_df['timestamp'].min()} to {combined_df['timestamp'].max()}")
print(f"  Duration: {combined_df['timestamp'].max() - combined_df['timestamp'].min()}")
print(f"\nMemory usage: {combined_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

## 7. Save Combined Dataset

```python
# Save to CSV
output_path = '5s_data/eth_btc_5s_l2_data.csv'
combined_df.to_csv(output_path, index=False)

print(f"✓ Combined dataset saved to: {output_path}")
print(f"  Shape: {combined_df.shape}")
print(f"  Columns: {len(combined_df.columns)} (1 timestamp + {len(eth_cols)} ETH + {len(btc_cols)} BTC)")
print(f"  File size: {combined_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```
